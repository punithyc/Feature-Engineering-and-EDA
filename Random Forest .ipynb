```python
import pandas as pd

import numpy as np

# Import libraries for data visualization
import matplotlib.pyplot as plt

import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("Titanic.csv")
df.shape
```

```python
df.head()
```

```python
#checking for null data

NAs = pd.concat([df.isnull().sum()], axis=1, keys=['null'])
NAs[NAs.sum(axis=1)>0]
```

```python
#Filling null age values with the mean

df["Age"]=df["Age"].fillna(df["Age"].mean())
```

```python
#Filling null embarked values with the mode

df["Embarked"]=df["Embarked"].fillna(df["Embarked"].mode()[0])
```

```python
#Filling null Cabin age values with the mean

df["Cabin"]=df["Cabin"].fillna(df["Cabin"].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.drop(['Cabin','Ticket','Name'], axis=1, inplace=True)
```

```python
df
```

```python
#Getting dummies from all the categorical variables

for col in df.dtypes[df.dtypes=='object'].index:
    for_dummy = df.pop(col)
    df = pd.concat([df, pd.get_dummies(for_dummy, prefix=col)], axis=1)
df.head()
```

```python
Y = df.pop('Survived')
```

```python
Y
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df, Y, test_size = 0.30)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
y_train
```

```python
y_pred = rf.predict(X_test)
```

```python
from sklearn.metrics import roc_curve, auc
false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_pred)
roc_curve = auc(false_positive_rate,true_positive_rate)
roc_curve  #goodness of the model
```
