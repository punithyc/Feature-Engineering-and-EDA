```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
data=load_iris()
```

```python
data
```

```python
data1=pd.DataFrame(data=data['data'],columns=data['feature_names'])
```

```python
data1
```

```python
data1['target']=data['target']
```

```python
data1
```

```python
x=data1.iloc[:,:-1]
y=data1['target']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
dc=DecisionTreeClassifier()
```

```python
dc.fit(x_train,y_train)
```

```python
x_predict=dc.predict(x_train)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(x_predict,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(dc,filled=True)
plt.show()
```

```python
dc1=DecisionTreeClassifier(ccp_alpha=0.24)
```

```python
dc1.fit(x_train,y_train)
```

```python
x_predict=dc1.predict(x_train)
```
