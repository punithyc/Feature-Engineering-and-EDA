```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
```

```python
x=np.array([[1,2],[2,2],[2,3],[8,7],[8,8],[25,80]])
```

```python
x
```

```python
plt.scatter(x[:,0],x[:,1])
plt.show()
```

```python
db=DBSCAN(eps=3,min_samples=2)
db.fit(x)
db.labels_
```

0 indicates 1st cluster and 1 indicates 2 nd cluster and -1 indicates noise point hereits creating 2 clusters because the min
_samples is 2 so it will check for the points having count of <=2 with its eps neighbourhood icluding it and it will create clusters with this core points and remaining will bw noise points

```python
db=DBSCAN(eps=3,min_samples=3)
db.fit(x)
db.labels_
```

0 indicates 1st cluster and -1 indicates noise point here its creating 1 cluster because the min _samples is 3 so it will check for the points having count of <=3 with its eps neighbourhood icluding it and it will create clusters with this core points and remaining will be noise points

```python
db=DBSCAN(eps=10,min_samples=2)
db.fit(x)
db.labels_
```

from the above the eps value is 10 so the radius of the circle is increased and so the neighbourhood is increased and the min_samples is 2 so it checks values in its neighbourhood and all the points are in the same eps radius so all the points are added into 1 cluster with 0 

```python
from sklearn.datasets import make_circles
```

```python
#creating a concentric circle dataset
x,_=make_circles(n_samples=500,factor=.5,noise=.03,random_state=4)
```

```python
#applying dbscan to dataset
dbscan=DBSCAN(eps=0.1,min_samples=5)
clusters=dbscan.fit_predict(x)

# plotting
plt.scatter(x[:,0],x[:,1],c=clusters,cmap='viridis',marker='o')
plt.title("dbscan clustering")
plt.xlabel("feature 0")
plt.ylabel("feature 1")
plt.show()
dbscan.labels_
```

from the we can see that dbscan is working with the circle data and in kmeans with this type of data we cant work and create clusters but dbscan will work with any type of data because it works ont density and not shapes
