```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=6*np.random.rand(100,1)-3
y=0.5*x**2+1.5*x+2+np.random.randn(100,1)
plt.scatter(x,y)
plt.xlabel('x dataset')
plt.ylabel('y dataset')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score
score=r2_score(lr.predict(x_test),y_test)
score
```

```python
plt.plot(x_train,lr.predict(x_train))
plt.scatter(x_train,y_train)
plt.xlabel("x")
plt.ylabel("y")
```

### polynomial regression

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pl=PolynomialFeatures(degree=2,include_bias=True)
x_train_pl=pl.fit_transform(x_train)
x_test_pl=pl.transform(x_test)
```

```python
x_train_pl
#,x0,x1,x1*2
```

```python
lr.fit(x_train_pl,y_train)
score=r2_score(lr.predict(x_test_pl),y_test)
score
```

```python
print(lr.coef_)
```

```python
print(lr.intercept_)
```

```python
plt.scatter(x_train,lr.predict(x_train_pl))
plt.scatter(x_train,y_train)
```

```python
pl=PolynomialFeatures(degree=3,include_bias=True)
x_train_pl=pl.fit_transform(x_train)
x_test_pl=pl.transform(x_test)
```

```python
lr.fit(x_train_pl,y_train)
score=r2_score(lr.predict(x_test_pl),y_test)
score
```

```python
plt.scatter(x_train,lr.predict(x_train_pl))
plt.scatter(x_train,y_train)
```
